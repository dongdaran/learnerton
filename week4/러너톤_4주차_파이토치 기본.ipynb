* .T transform method

```python
tensor_a = torch.tensor([[1,2,3],
                         [4,5,6]])
tensor_b = torch.tensor([[-1,-2,-3],
                         [-4,-5,-6]])
```

```python
torch.matmul(tensor_a, tensor_b.T)
```

```python
tensor_axis_max = torch.max(tensor_a, axis = 1)
```

미분(역전파)
-tensor_x 값이 tensor_z 값에 미치는 영향도(미분값)

```python
tensor_x = torch.tensor([[1., 2., 3.],
                        [4., 5., 6.]], requires_grad= True)

tensor_y = torch.tensor([[7., 8., 9.],
                        [10., 11., 12.]], requires_grad= True)
```

```python
tensor_z = torch.square(tensor_x + tensor_y)
```

```python
tensor_z.sum().backward()
```

```python
gradient_zx = tensor_x.grad
```

```python
print(np.round(gradient_zx, 2))
```

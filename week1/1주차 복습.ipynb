#Python 기초

```python
course_list = ["Python Basic", "Java Basic",
             "Python Machine Learning",
             "Java Web Development",
             "Embedded Programming with C++",
             "Data Science with Python",
             "Python Web Development",
             "Game Programming with Unity"]
```

```python
print(*enumerate(course_list))
# enumerate는 포인터 형식으로 반환
```

```python
for index, course in enumerate(course_list):
    print(index, course)
```

```python
# dict에서 key와 value는 dict_key, dict_value 자료형이기에 list로 바꿔줘야함
course_dict = {
  "name": "Python Programming",
  "level":  "entry",
  "Price": 49900,
  "Rating": 4.98
}
```

```python
print(type(course_dict.keys()))
print(type(course_dict.values()))

key_list = list(course_dict.keys())
value_list = list(course_dict.values())

print(key_list, type(key_list))
print(value_list, type(value_list))
```

```python
#.get method는 해당하는 값이 없으면 None을 반환
# list엔 없다네요^^
print(course_dict.get("including"))
print(course_dict.get("name"))
```

#NumPy

```python
import numpy as np
```

```python
data = [10, 20, 30]
arr = np.array(data)

print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python
data = [[10, 20, 30],[40,50,60]]
arr = np.array(data)

print(arr,"\n")
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python
data = [[10, 20, 30],[40,50,60]]
arr = np.array(object=data, dtype = float)

print(arr,"\n")
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python
data = [[[1,2,3],
          [4,5,6]],
        [[3,2,1],
         [6,5,4]]]

arr = np.array(data)

print(arr, '\n')
print(arr.dtype)
print(arr.shape)
print(arr.size)
```

```python
# zeros와 같은 함수는 argument를 tuple로 제공해야함
np.empty((3,5,6))
```

```python
# ndarray인자를 안넣어도 된다
data = [[1,2,3],[4,5,6]]
one_array = np.ones_like(data)

print(data, type(data))
print()
print(one_array, type(one_array))
```

```python
# 특정 타입 변경 method astype
origin = np.arange(1,2,0.2)
change = origin.astype(int)

print(origin)
print(change)
```

```python
origin = np.arange(20)
print(origin)
```

```python
change1 = origin.reshape(2,2,5)
print(change1)

change2 = np.reshape(origin,(-1,2,5))
print(change2)
```

```python
print(change)
```

```python
change = origin.reshape(2,2,5)

print(change.sum(axis=0))
print()
print(change.sum(axis=1))
print()
print(change.sum(axis=2))
```

```python
print(change)
print("-------------------------")
print(np.sort(change, axis=0)[::-1,:,:])
print("-------------------------")
print(np.sort(change, axis=0)[:,::-1,:])
print("-------------------------")
print(np.sort(change, axis=0)[:,:,::-1])
print("-------------------------")
```

##Series

```python
import pandas as pd
```

```python
year = [2019, 2020, 2021, 2022]
result = pd.Series(year)

print(type(result))
print(result)
```

```python
year=[2019,2020,2021,2022]
idx=['a','b','c','d']

result = pd.Series(data= year, index=idx, name ='Year')
print(result)
```

```python
score = {'kim': 80,
         'han':85,
         'Lee':90,
         'Choi':95}

result = pd.Series(data=score, name='score')
print(result)
```

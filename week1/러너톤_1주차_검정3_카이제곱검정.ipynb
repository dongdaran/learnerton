#1. one-way x-test

* 실제 ~ 기대도수 차이 검정

* 검정과정 : 빈도교차표 -> 기대도수 -> 카이제곱 통계량 산출

---
* 문제
  * 기존 교통사고의 심각여부 남성의 경우 20,30,40,50,60 대 각각의 비율이 10%, 20%, 20%, 30%, 30%로 나타남
* 가설
  * H0 : 기존이 맞다
  * H1 : 기존이 아니다

```python
serious_df = df.loc[df['serious_yn']=='Y']
```

1. 빈도 교차표

```python
X = pd.crosstab(serious_df['sex'], serious_df['age'], margins=True)
X
```

```python
print(X.values)
```

```python
m = X.values[1, :5]
m
```

---
다른방법

```python
serious_df_male = serious_df.loc[serious_df['sex']=='M']
```

```python
XX = pd.crosstab(serious_df_male['age'], serious_df_male['serious_yn'], margins=True)
XX.values
```

```python
mm=XX.values[:5,0]
print(mm)
```

----
2. 기대도수

```python
# 기존 사실
Pr = np.array([0.1,0.2,0.2,0.3,0.2])
```

```python
n = X.values[1,5]
n
```

```python
E = Pr*n
E
```

3. 검증

```python
stats.chisquare(f_obs=m, f_exp=E)
```

#2. two-way x-test

* 두개 범주형 변수 간 연관성 검정

* 검정과정 : 빈도교차표 -> 기대도수 -> 카이제곱 통계량 산출

---
* 문제
  * 교통사고 심각여부 ~ 성별 연관성 유무
* 가설
  * H0 : 연관성 없다
  * H1 : 연관성 있다

1. 빈도교차표

```python
X = pd.crosstab(df['sex'], df['serious_yn'], margins=True)
X
```

2. 기대도수 & 3. 카이제곱 검정
two-way x-test는 2,3을 동시에 해줌

```python
m = X.values[:2,:2]
```

```python
stats.chi2_contingency(m)
```

```python
data = [10,20,30]
arr = np.array(data)

print("1D ndarray : ")
print(arr)
print("type : ", type(arr))
print("shape : {}, size : {}, dtype : {}, dimension: {}".format(arr.shape, arr.size, arr.dtype, arr.ndim))
```

```python
data = [[1,2,3], [4,5,6]]
arr = np.array(data)

print("2D ndarray : ")
print(arr)
print("type : ", type(arr))
print("shape : {}, size : {}, dtype : {}, dimension: {}".format(arr.shape, arr.size, arr.dtype, arr.ndim))
```

```python
origin = np.arange(1,2, 0.2)
change = origin.astype(int)

print("<origin>")
print(origin)
print()
print("<change>")
print(change)
```

```python
origin =np.arange(15)
reshape_arr1 = np.reshape(origin, (3,5))
reshape_arr2 = origin.reshape((3,5))
print("<origin>")
print(origin)
print()
print("<reshape_arr1>")
print(reshape_arr1)
print()
print("<reshape_arr2>")
print(reshape_arr2)
```

```python
origin =np.arange(15)
reshape_arr = np.reshape(origin, (-1,1))

print("<origin>")
print(origin)
print()
print("<reshape_arr>")
print(reshape_arr)
print()
```

```python
A = np.arange(1,10).reshape(3,3)
B = np.sort(np.arange(1,10))[::-1].reshape(3,3)

print("A--")
print(A)
print()
print("B--")
print(B)
print()
print("result of A>B--")
print(A>B)
```

```python
sample = np.arange(16).reshape(4,2,2)
print(sample)
print()
print("index (1,1,1) :" ,sample[1,1,1])
```

```python
sample = np.arange(1,17).reshape(4,4)
print("<all>")
print(sample)
print()
print("<[:,1:-1]>")
print(sample[:,1:-1])
print()
print("<[:-1,2]>")
print(sample[:-1,2])
print()
print("<[:,-1]>")
print(sample[:,-1])
```

```python
name = np.array(['John', 'Samuel', 'Kate', 'Mike', 'Sarah'])
score = np.array([78, 84, 96, 88, 82])

sort_indexes = np.argsort(score)[::-1]
print("sort indexes : ", sort_indexes)
print("sort values : ", name[sort_indexes])
```

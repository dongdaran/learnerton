```python
import pandas as pd
import numpy as np
import json

# 유클리드 거리 계산 함수
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)

# 드라이버 데이터 읽기
def read_drivers(filename):
    drivers_df = pd.read_csv(filename)
    drivers = []
    for _, row in drivers_df.iterrows():
        drivers.append({
            'id': row['DeliveryID'],
            'current_x': row['CurrentX'],
            'current_y': row['CurrentY'],
            'capacity': row['Capacity'],
            'completion_time': 0.0,
            'route': []
        })
    return drivers

# 주문 데이터 읽기
def read_orders(filename):
    orders_df = pd.read_csv(filename)
    orders = []
    for _, row in orders_df.iterrows():
        orders.append({
            'id': row['OrderID'],
            'pickup_x': row['PickupX'],
            'pickup_y': row['PickupY'],
            'delivery_x': row['DeliveryX'],
            'delivery_y': row['DeliveryY'],
            'size': row['OrderSize']
        })
    return orders

# 경로 최적화를 위한 그리디 알고리즘
def optimize_route(driver, orders):
    current_x, current_y, capacity = driver['current_x'], driver['current_y'], driver['capacity']

    # 경로를 최적화하는 리스트 (픽업과 배달을 모두 고려)
    route = []
    remaining_capacity = capacity
    remaining_orders = orders.copy()

    # 각 주문에 대해 가장 가까운 픽업과 배달을 처리
    while remaining_orders:
        min_distance = float('inf')
        closest_order = None

        # 픽업 또는 배달을 할 수 있는 주문을 찾는다
        for order in remaining_orders:
            # 픽업과 배달 위치를 차례로 처리
            for order_type, order_x, order_y in [('P', order['pickup_x'], order['pickup_y']),
                                                  ('D', order['delivery_x'], order['delivery_y'])]:
                # 용량을 고려해서 배달을 할 수 있는 경우에만 처리
                if order_type == 'P' and order['size'] <= remaining_capacity:
                    dist = euclidean_distance(current_x, current_y, order_x, order_y)
                    if dist < min_distance:
                        min_distance = dist
                        closest_order = (order_type, order)
                elif order_type == 'D' and order['size'] <= remaining_capacity:
                    dist = euclidean_distance(current_x, current_y, order_x, order_y)
                    if dist < min_distance:
                        min_distance = dist
                        closest_order = (order_type, order)

        if closest_order:
            order_type, order = closest_order

            # 픽업인 경우 용량을 차감하고 경로에 추가
            if order_type == 'P':
                route.append(f"P:{order['id']}")
                remaining_capacity -= order['size']
                current_x, current_y = order['pickup_x'], order['pickup_y']
            # 배달인 경우 용량을 회복하고 경로에 추가
            elif order_type == 'D':
                route.append(f"D:{order['id']}")
                remaining_capacity += order['size']
                current_x, current_y = order['delivery_x'], order['delivery_y']

            # 처리된 주문은 남은 주문 리스트에서 제거
            remaining_orders.remove(order)

    return route

# 메인 함수
def main():
    drivers = read_drivers('deliver.csv')
    orders = read_orders('orders.csv')

    for driver in drivers:
        # 각 드라이버에 대해 경로를 최적화
        optimized_route = optimize_route(driver, orders)
        driver['route'] = optimized_route

    # 결과를 JSON 형식으로 저장
    output = {driver['id']: driver['route'] for driver in drivers}
    with open('output.json', 'w') as f:
        json.dump(output, f, indent=2)

if __name__ == "__main__":
    main()
```

```python
import csv
import math
import json
from datetime import timedelta
from hiddenFiles import codepresso_util as cu


def load_orders(filename="orders.csv"):
  orders = []
  with open(filename, newline="", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for row in reader:
      row["PickupX"] = int(row["PickupX"])
      row["PickupY"] = int(row["PickupY"])
      row["DeliveryX"] = int(row["DeliveryX"])
      row["DeliveryY"] = int(row["DeliveryY"])
      row["OrderSize"] = int(row["OrderSize"])
      row["pickup"] = (row["PickupX"], row["PickupY"])
      row["delivery"] = (row["DeliveryX"], row["DeliveryY"])
      orders.append(row)
  # 주문은 OrderID 순으로 처리
  orders.sort(key=lambda r: r["OrderID"])
  return orders

def load_deliveries(filename="deliver.csv"):
  deliveries = {}
  with open(filename, newline="", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    for row in reader:
      row["CurrentX"] = int(row["CurrentX"])
      row["CurrentY"] = int(row["CurrentY"])
      row["Capacity"] = int(row["Capacity"])
      row["position"] = (row["CurrentX"], row["CurrentY"])
      row["available_time"] = 0  # 시작 시각을 0분으로 가정
      deliveries[row["DeliveryID"]] = row
  return deliveries

def assign_orders(orders, deliveries):
  # $ANNOTATION_1
  import numpy as np

  route = {}
  deliveries_initial = deliveries.copy()

  for order in orders:
    min_delivery_time = np.inf
    best_delivery = None

    for delivery in deliveries.values():
      if delivery['Capacity'] < order['OrderSize']:
        continue
      c_p = ((order["PickupX"]-delivery["CurrentX"])**2+(order["PickupY"]-delivery["CurrentY"])**2)**(1/2)
      p_o = ((order["PickupX"]-order["DeliveryX"])**2+(order["PickupY"]-order["DeliveryY"])**2)**(1/2)
      total_delivery_time = int(c_p + p_o)

      if total_delivery_time+delivery["available_time"] < min_delivery_time:
        best_delivery = delivery
        min_delivery_time = total_delivery_time+delivery["available_time"]

    best_delivery["CurrentX"]=order["DeliveryX"]
    best_delivery["CurrentY"]=order["DeliveryY"]
    best_delivery["available_time"] =min_delivery_time
    if route.get(best_delivery['DeliveryID']) == None:
      route[best_delivery['DeliveryID']] =[]
      route[best_delivery['DeliveryID']].append("P:{}".format(order["OrderID"]))
      route[best_delivery['DeliveryID']].append("D:{}".format(order["OrderID"]))
    else:
      route[best_delivery['DeliveryID']].append("P:{}".format(order["OrderID"]))
      route[best_delivery['DeliveryID']].append("D:{}".format(order["OrderID"]))

  for deliver_id, indi_route in route.items():
    deliver=deliveries[deliver_id]
    deliver['position']=deliveries_initial[deliver_id]['position']
    sort_route =[]
    for n in range(len(indi_route)):
      indi_route[n]=indi_route[n][3:]

    indi_route=set(indi_route)
    indi_route_len=len(indi_route)
    for j in range(indi_route_len):
      best=None
      min_time = np.inf
      for i in indi_route:
        int_i = int(i)
        c_p = ((orders[int_i]["PickupX"]-deliver["CurrentX"])**2+(orders[int_i]["PickupY"]-deliver["CurrentY"])**2)**(1/2)
        p_o = ((orders[int_i]["PickupX"]-orders[int_i]["DeliveryX"])**2+(orders[int_i]["PickupY"]-orders[int_i]["DeliveryY"])**2)**(1/2)
        total_time = c_p + p_o

        if total_time < min_time:
          best = i
        sort_route.append("P:O{}".format(best))
        sort_route.append("D:O{}".format(best))
      route[deliver_id] = sort_route

  return route


# $ANNOTATION_2
if __name__ == "__main__":
  orders = load_orders("orders.csv")
  deliveries = load_deliveries("deliver.csv")
  actions_by_delivery = assign_orders(orders, deliveries)
  cu.write_submission_json(actions_by_delivery)
  cu.validate()
```

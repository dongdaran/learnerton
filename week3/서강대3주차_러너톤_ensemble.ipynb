```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

#데이터 준비

```python
cancer = load_breast_cancer()
x_train, x_test, y_train, y_test = train_test_split(cancer.data,
													cancer.target,
													test_size=0.3,
													random_state=12)
```

#Voting
VotingClassifier (estimators=[('모델',변수),('모델',변수)...], voting = 'soft')

.fit(x_train, y_train)

.predict(x_test)

```python
#앙상블 할 모델 설정
lr = LogisticRegression(random_state =12, max_iter =5000)
dt = DecisionTreeClassifier(random_state =12)

#Voting 진행
voting = VotingClassifier(estimators=[('LR', lr),('DT',dt)],voting='soft')
voting.fit(x_train, y_train)
pred = voting.predict(x_test)
```

#Score

```python
print("Accuracy: {0:3f}".format(accuracy_score(y_test, pred)))
```
